### Comienzo a analizar los tipos para el analisis del tipo dragon

```python
value_counts_col1 = pokefilt["Type 1"].value_counts()
value_counts_col2 = pokefilt["Type 2"].value_counts()
dfTipos = pd.concat([value_counts_col1, value_counts_col2,value_counts_col1+value_counts_col2], axis=1)
dfTipos.columns = ["Type 1", "Type 2", "Total"]
dfTipos
```

### Despues de analizar, me dispongo a sacar graficas para explicar mis conclusiones

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=dfTipos.index, y=dfTipos["Total"])
plt.xticks(rotation=-45)
plt.title('Conteo Total de Tipos de Pokémon')
plt.xlabel('Tipo de Pokémon')
plt.ylabel('Conteo');
```

```python
plt.figure(figsize=(8, 8))
dfTipos['Total'].plot(kind='pie', labels=dfTipos.index, autopct='%1.0f%%', legend=False)
plt.title("Porcentaje de Tipos de Pokémon")
plt.ylabel("")
```
